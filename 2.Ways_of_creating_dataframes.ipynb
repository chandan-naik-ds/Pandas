```python
# 3rd waay is using python dictonaries 
import pandas as pd

weather = {
    'day': ['1-1-2017', '1-2-2017', '1-3-2017'],
    'temperature': [32, 35, 34],
    'event': ['sunny', 'rainy', 'snow']
}

df = pd.DataFrame(weather)
df
```

```python
# 4th way is by creating python tuples

wheather = [
    ('1-1-2017', 32, 'Rain'),
    ('1-2-2017', 35, 'sunny'),
    ('1-3-2017', 36, 'snow')
]

df = pd.DataFrame(wheather, columns= ["day", "temperature", "event"])
df
```

```python
# another way using python dictonaries

wheather = [
    {'day':"1-1-2017",'temperature': 32,'event':"Rain"},
    {'day':"1-2-2017",'temperature': 35,'event':"sunny"},
    {'day':"1-1-2017",'temperature': 34,'event':"snow"}
]

df = pd.DataFrame(wheather)
df
```
